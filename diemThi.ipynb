# Xử lý dữ liệu
Nếu dữ liệu ở cột Diem_HP > 10 => Chuyển về 10

```python
diemThi["Diem_HP"] = diemThi["Diem_HP"].apply(lambda x: 10 if x > 10 else x)
diemThi.to_pickle("diemThi.pkl")
```

```python
# Xoá các dòng dữ liệu thiếu điểm học phần
diemThi[["MaMH", "MaMH_tt"]]
MaMH = diemThi["MaMH"].unique()[0]
df_diemThiMH = diemThi.loc[diemThi["MaMH"] == MaMH]
```

```python
df_diemThiMH.groupby(["NamHoc","HocKy"]).mean().reset_index()
df_diemThiMH
```

```python
fig = px.histogram(df_diemThiMH, x = "Diem_HP",
                   color="NamHoc",
                   title= "Thông tin điểm thi qua từng năm của môn học")
fig
```

```python
fig = px.histogram(df_diemThiMH, x = "NamHoc", y = "HocKy",
                   color="HocKy", text_auto=True,
                   barmode="group",
                   title= "Thông tin điểm thi qua từng năm của môn học")
fig
```
