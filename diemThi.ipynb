# Phân tích các đầu mã môn học trong toàn bộ môn

```python
# Chia cột 'column_name' thành 'letters' và 'numbers'
diemThi['MaMH_tiento'] = diemThi['MaMH'].str.extract(r'([A-Za-z]+)', expand=False)
diemThi['MaMH_hauto'] = diemThi['MaMH'].str.extract(r'(\d+)', expand=False)
diemThi['MaMH_tiento'].to_pickle("MaMH_tiento.pkl")
```

```python
# MaMH_tiento = pd.read_pickle("MaMH_tiento.pkl")
diemThi_tt = diemThi.groupby("MaMH_tiento").size().reset_index()
mapping = ['MaMH_tiento', "Số lượng"]
diemThi_tt.columns = ['MaMH_tiento', "SoLuong"]
diemThi_tt.sort_values(by = "SoLuong", ascending=False, inplace=True)
diemThi_tt.to_pickle("MaMH_tt.pkl")
```
